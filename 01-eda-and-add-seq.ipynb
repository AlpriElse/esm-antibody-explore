```python
from Bio.PDB import PDBParser, PDBIO, Select

class CASelect(Select):
    def accept_atom(self, atom):
        return atom.get_id() == 'CA'

def extract_ca_atoms(pdb_file, output_file):
    parser = PDBParser(QUIET=True)
    structure = parser.get_structure('protein', pdb_file)
    
    io = PDBIO()
    io.set_structure(structure)
    io.save(output_file, CASelect())

# Example usage:
pdb_file = 'data/all-nano-structures/8zer.pdb'
output_file = './data/test-Ca-8zer.pdb'
extract_ca_atoms(pdb_file, output_file)
```

# Getting Sequences out of PDB files

```python
from Bio.PDB.Polypeptide import PPBuilder, three_to_index, index_to_one, is_aa

def chain_to_seq(chain):
    try:
        residues = [r.get_resname() for r in chain.get_residues()]

        output = ''
        for r in residues:
            if not is_aa(r):
                continue
            output += index_to_one(three_to_index(r))
        return output
    except Exception as e:
        print('something went wrong converting chain to seq', e)
        return None
```

```python
import os 
BASE_PATH = 'data/all-nano-structures'
def read_pdb_structure(structure_id):
    full_path = f'{BASE_PATH}/{structure_id}.pdb'
    
    if not os.path.isfile(full_path):
        return None
    
    structure = parser.get_structure(structure_id, full_path)
    return structure
```

```python
def get_sequence_from_row(row):
    structure_id = row['pdb']
    chain_id = row['Hchain']
    
    if not structure_id or not chain_id:
        print('sequence_id or chain_id is None:', row)
        return None

    pdb_structure = read_pdb_structure(structure_id)
    
    if len(pdb_structure) != 1:
        print('pdb structure does not have exactly 1 model:', row)
        return None
    
    return chain_to_seq(pdb_structure[0][chain_id])
```
